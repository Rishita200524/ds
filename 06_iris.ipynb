```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv('iris.csv')
```

```python
# Display first rows
print(df.head())
```

```python
# Subset: only Setosa species
setosa = df[df['species'] == 'setosa']
print(setosa.head())
```

```python
# Subset: select specific columns
subset_cols = df[['sepal_length', 'sepal_width']]
print(subset_cols.head())
```

```python
# Subset using conditions
subset_cond = df[df['sepal_length'] > 5.0]
print(subset_cond.head())
```

```python
# Create another dataframe (example)
df2 = df[['sepal_length', 'species']]
```

```python
# Merge on common column
merged = pd.merge(df, df2, on='sepal_length')
```

```python
print(merged.head())
```

```python
# Sort by sepal_length
sorted_df = df.sort_values(by='sepal_length')
print(sorted_df.head())
```

```python
# Sort by multiple columns
sorted_multi = df.sort_values(by=['species', 'petal_length'])
print(sorted_multi.head())
```

```python
print(df.shape)   # (rows, columns)
```

```python
pivot_table = df.pivot_table(values='sepal_length', index='species', aggfunc='mean')
print(pivot_table)
```

```python
melted = pd.melt(df, id_vars=['species'])
print(melted.head())
```

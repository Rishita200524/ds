```python
import pandas as pd
```

```python
df = pd.read_csv("heart.csv")
print(df.head())
```

```python
#2. Create Data Subsets
#column subsset
subset_cols = df[['age', 'sex', 'cp', 'chol', 'output']]
print(subset_cols.head())
```

```python
#Row Subset
subset_rows = df.iloc[:10]
print(subset_rows)
```

```python
#Merge Data
merged_df = pd.merge(df, subset_cols,
                     on='age',
                     how='inner',
                     suffixes=('_left', '_right'))
```

```python
print(merged_df.head())
```

```python
#sort data
sorted_df = df.sort_values(by='chol', ascending=False)
print(sorted_df.head())
```

```python
#Transpose Data
transposed_df = df.head().T
print(transposed_df)
```

```python
#6. Shape and Reshape Data
#Shape of Dataset
print(df.shape)
```

```python
#Reshape using Pivot Table
pivot_df = df.pivot_table(values='chol',
                          index='age',
                          columns='sex',
                          aggfunc='mean')
```

```python
print(pivot_df)
```

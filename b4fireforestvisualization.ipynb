```python
plt.scatter(df['month'], df['rain'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
```

```python
plt.xlabel('Month')
plt.ylabel('rain')
plt.show()
```

```python
#Scatter Plot of month and wind
plt.scatter(df['month'], df['wind'])
# Adding Title to the Plot
plt.title("Scatter Plot")
# Setting the X and Y labels
plt.xlabel('Month')
plt.ylabel('wind')
plt.show()
```

```python
plt.bar(df['day'], df['temp'])
plt.title("Bar Chart")
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.show()
```

```python
plt.bar(df['month'], df['temp'])
plt.title("Bar Chart")
plt.xlabel('Month')
plt.ylabel('Temprature')
plt.show()
```

```python
#Histogram
plt.hist(df['RH'])
plt.title("Histogram")
plt.show()
```

```python
#Box Plot
df
df.boxplot(by ='day', column =['temp'], grid = False)
<Axes: title={'center': 'temp'}, xlabel='day'>
```

```python
#Scatter Chart
sn.scatterplot(x='month', y='temp', data=df,hue='day')
plt.title('Scatter Chart')
plt.show()
```

```python
# Line Plot
sn.lineplot(x="month", y="temp", data=df)
plt.title('Line Chart')
plt.show()
```

```python
# Keep only numeric columns
numeric_df =
 df.select_dtypes(include=['number'])
# Calculate the correlation matrix
corr_matrix =numeric_df.corr()
# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
